```python
covid19_genome = ''

covid19_skew = [0]
for i in range(len(covid19_genome)):
    if covid19_genome[i] == 'C': covid19_skew += [covid19_skew[i]-1]
    elif covid19_genome[i] == 'G': covid19_skew += [covid19_skew[i]+1]
    else: covid19_skew += [covid19_skew[i]]

print('minimum is at: '+str(covid19_skew.index(min(covid19_skew)))+' position')

import matplotlib.pyplot as plt
%matplotlib inline
x = [i for i in range(len(covid19_genome)+1)]
plt.plot(x, covid19_skew, 'black')
plt.xlabel('Position')
plt.ylabel('Skew')
plt.title('Skew diagram of COVID-19') 
plt.show()
```

В качестве второго 500 позиций от минимума(218 позиция).

```python
cvd_skew_min = covid19_skew.index(min(covid19_skew))
covid19_oric1 = covid19_genome[:500]
covid19_oric2 = covid19_genome[cvd_skew_min:cvd_skew_min+500]


def hamming_dist(line1,line2):
    if len(line1) != len(line2): return None
    dist = 0
    for i in range(len(line1)):
        if line1[i] != line2[i]: dist += 1
    return dist


def neighbours(line,k,d):
    if d == 0: return line
    if k == 1: return ['A','C','G','T']
    prefix_neighbours = neighbours(line[:k-1],k-1,d)
    neighbourhood = []
    for pre_line in prefix_neighbours:
        if hamming_dist(pre_line,line[:k-1]) < d:
            neighbourhood += [pre_line + 'A']
            neighbourhood += [pre_line + 'C']
            neighbourhood += [pre_line + 'G']
            neighbourhood += [pre_line + 'T']
        else:
            neighbourhood += [pre_line + line[k-1]]
    return neighbourhood

cvd19_oric1_nbh_8_0 = []
cvd19_oric2_nbh_8_0 = []
cvd19_oric1_nbh_8_1 = []
cvd19_oric2_nbh_8_1 = []
cvd19_oric1_nbh_8_2 = []
cvd19_oric2_nbh_8_2 = []
cvd19_oric1_nbh_9_0 = []
cvd19_oric2_nbh_9_0 = []
cvd19_oric1_nbh_9_1 = []
cvd19_oric2_nbh_9_1 = []
cvd19_oric1_nbh_9_2 = []
cvd19_oric2_nbh_9_2 = []
for i in range(491):
    cvd19_oric1_nbh_8_0 += [covid19_oric1[i:i+8]]
    cvd19_oric2_nbh_8_0 += [covid19_oric2[i:i+8]]
    cvd19_oric1_nbh_8_1 += neighbours(covid19_oric1[i:i+8],8,1)
    cvd19_oric2_nbh_8_1 += neighbours(covid19_oric2[i:i+8],8,1)
    cvd19_oric1_nbh_8_2 += neighbours(covid19_oric1[i:i+8],8,2)
    cvd19_oric2_nbh_8_2 += neighbours(covid19_oric2[i:i+8],8,2)
    cvd19_oric1_nbh_9_0 += [covid19_oric1[i:i+9]]
    cvd19_oric2_nbh_9_0 += [covid19_oric1[i:i+9]]
    cvd19_oric1_nbh_9_1 += neighbours(covid19_oric1[i:i+9],9,1)
    cvd19_oric2_nbh_9_1 += neighbours(covid19_oric2[i:i+9],9,1)
    cvd19_oric1_nbh_9_2 += neighbours(covid19_oric1[i:i+9],9,2)
    cvd19_oric2_nbh_9_2 += neighbours(covid19_oric2[i:i+9],9,2)


cvd19_oric1_nbh_8_0 += [covid19_oric1[492:500]]
cvd19_oric2_nbh_8_0 += [covid19_oric2[492:500]]
cvd19_oric1_nbh_8_1 += neighbours(covid19_oric1[492:500],8,1)
cvd19_oric2_nbh_8_1 += neighbours(covid19_oric2[492:500],8,1)
cvd19_oric1_nbh_8_2 += neighbours(covid19_oric1[492:500],8,2)
cvd19_oric2_nbh_8_2 += neighbours(covid19_oric2[492:500],8,2)


count8_1_0 = {}
for n in cvd19_oric1_nbh_8_0: count8_1_0[n] = 0

count8_2_0 = {}
for n in cvd19_oric2_nbh_8_0: count8_2_0[n] = 0

count8_1_1 = {}
for n in cvd19_oric1_nbh_8_1: count8_1_1[n] = 0

count8_2_1 = {}
for n in cvd19_oric2_nbh_8_1: count8_2_1[n] = 0

count8_1_2 = {}
for n in cvd19_oric1_nbh_8_2: count8_1_2[n] = 0

count8_2_2 = {}
for n in cvd19_oric2_nbh_8_2: count8_2_2[n] = 0

count9_1_0 = {}
for n in cvd19_oric1_nbh_9_0: count9_1_0[n] = 0

count9_2_0 = {}
for n in cvd19_oric2_nbh_9_0: count9_2_0[n] = 0

count9_1_1 = {}
for n in cvd19_oric1_nbh_9_1: count9_1_1[n] = 0

count9_2_1 = {}
for n in cvd19_oric2_nbh_9_1: count9_2_1[n] = 0

count9_1_2 = {}
for n in cvd19_oric1_nbh_9_2: count9_1_2[n] = 0

count9_2_2 = {}
for n in cvd19_oric2_nbh_9_2: count9_2_2[n] = 0


for i in range(491):
    read1 = covid19_oric1[i:i+8]
    read2 = covid19_oric2[i:i+8]
    for c in count8_1_0:
        if hamming_dist(read1,c) == 0: count8_1_0[c] += 1
    for c in count8_2_0:
        if hamming_dist(read2,c) == 0: count8_2_0[c] += 1
    for c in count8_1_1:
        if hamming_dist(read1,c) <= 1: count8_1_1[c] += 1
    for c in count8_2_1:
        if hamming_dist(read2,c) <= 1: count8_2_1[c] += 1
    for c in count8_1_2:
        if hamming_dist(read1,c) <= 2: count8_1_2[c] += 1
    for c in count8_2_2:
        if hamming_dist(read1,c) <= 2: count8_2_2[c] += 1

    read1 += covid19_oric1[i+8]
    read2 += covid19_oric2[i+8]

    for c in count9_1_0:
        if hamming_dist(read1,c) == 0: count9_1_0[c] += 1
    for c in count9_2_0:
        if hamming_dist(read2,c) == 0: count9_2_0[c] += 1
    for c in count9_1_1:
        if hamming_dist(read1,c) <= 1: count9_1_1[c] += 1
    for c in count9_2_1:
        if hamming_dist(read2,c) <= 1: count9_2_1[c] += 1
    for c in count9_1_2:
        if hamming_dist(read1,c) <= 2: count9_1_2[c] += 1
    for c in count9_2_2:
        if hamming_dist(read2,c) <= 2: count9_2_2[c] += 1

for c in count9_1_0:
    if hamming_dist(covid19_oric1[491:500],c) == 0: count9_1_0[c] += 1
for c in count9_2_0:
    if hamming_dist(covid19_oric2[491:500],c) == 0: count9_2_0[c] += 1
for c in count9_1_1:
    if hamming_dist(covid19_oric1[491:500],c) <= 1: count9_1_1[c] += 1
for c in count9_2_1:
    if hamming_dist(covid19_oric2[491:500],c) <= 1: count9_2_1[c] += 1
for c in count9_1_2:
    if hamming_dist(covid19_oric1[491:500],c) <= 2: count9_1_2[c] += 1
for c in count9_2_2:
    if hamming_dist(covid19_oric2[491:500],c) <= 2: count9_2_2[c] += 1



print('8mer oric1 d=0')
out8_1_0 = []
m = max(count8_1_0.values())
for c in count8_1_0:
    if count8_1_0[c] == m: out8_1_0 += [c]
print(out8_1_0,m,'\n')

print('8mer oric2 d=0')
out8_2_0 = []
m = max(count8_2_0.values())
for c in count8_2_0:
    if count8_2_0[c] == m: out8_2_0 += [c]
print(out8_2_0,m,'\n')

print('8mer oric1 d=1')
out8_1_1 = []
m = max(count8_1_1.values())
for c in count8_1_1:
    if count8_1_1[c] == m: out8_1_1 += [c]
print(out8_1_1,m)

print('8mer oric2 d=1')
out8_2_1 = []
m = max(count8_2_1.values())
for c in count8_2_1:
    if count8_2_1[c] == m: out8_2_1 += [c]
print(out8_2_1,m,'\n')

print('8mer oric1 d=2')
out8_1_2 = []
m = max(count8_1_2.values())
for c in count8_1_2:
    if count8_1_2[c] == m: out8_1_2 += [c]
print(out8_1_2,m,'\n')

print('8mer oric2 d=2')
out8_2_2 = []
m = max(count8_2_2.values())
for c in count8_2_2:
    if count8_2_2[c] == m: out8_2_2 += [c]
print(out8_2_2,m,'\n')

print('9mer oric1 d=0')
out9_1_0 = []
m = max(count9_1_0.values())
for c in count9_1_0:
    if count9_1_0[c] == m: out9_1_0 += [c]
print(out9_1_0,m,'\n')

print('9mer oric2 d=0')
out9_2_0 = []
m = max(count9_2_0.values())
for c in count9_2_0:
    if count9_2_0[c] == m: out9_2_0 += [c]
print(out9_2_0,m,'\n')

print('9mer oric1 d=1')
out9_1_1 = []
m = max(count9_1_1.values())
for c in count9_1_1:
    if count9_1_1[c] == m: out9_1_1 += [c]
print(out9_1_1,m,'\n')

print('9mer oric2 d=1')
out9_2_1 = []
m = max(count9_2_1.values())
for c in count9_2_1:
    if count9_2_1[c] == m: out9_2_1 += [c]
print(out9_2_1,m,'\n')

print('9mer oric1 d=2')
out9_1_2 = []
m = max(count9_1_2.values())
for c in count9_1_2:
    if count9_1_2[c] == m: out9_1_2 += [c]
print(out9_1_2,m,'\n')

print('9mer oric2 d=2')
out9_2_2 = []
m = max(count9_2_2.values())
for c in count9_2_2:
    if count9_2_2[c] == m: out9_2_2 += [c]
print(out9_2_2,m,'\n')
```
